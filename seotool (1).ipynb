```python
import gradio as gr
import requests
from bs4 import BeautifulSoup
import validators

def fetch_page(url):
    try:
        headers = {'User-Agent': 'Mozilla/5.0'}
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
        return BeautifulSoup(response.text, 'html.parser')
    except requests.exceptions.RequestException:
        return None

def check_meta_tags(soup):
    issues = []
    title = soup.find('title')
    meta_desc = soup.find('meta', attrs={'name': 'description'})

    if not title or not title.text.strip():
        issues.append("‚ùå Missing or empty <title> tag.")
    elif len(title.text.strip()) > 60:
        issues.append("‚ö†Ô∏è Title is too long (over 60 characters).")

    if not meta_desc or not meta_desc.get('content', '').strip():
        issues.append("‚ùå Missing or empty meta description.")
    elif len(meta_desc['content']) > 160:
        issues.append("‚ö†Ô∏è Meta description is too long (over 160 characters).")

    return issues

def check_broken_links(soup, url):
    broken_links = []
    base_url = '/'.join(url.split('/')[:3])
    for link in soup.find_all('a', href=True):
        href = link['href']
        full_url = href if validators.url(href) else base_url + href
        try:
            response = requests.head(full_url, timeout=5, allow_redirects=True)
            if response.status_code >= 400:
                broken_links.append(full_url)
        except requests.exceptions.RequestException:
            broken_links.append(full_url)
    return broken_links

def check_headings(soup):
    headings = [h.name for h in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])]
    issues = []
    if headings.count('h1') == 0:
        issues.append("‚ùå Missing H1 tag.")
    elif headings.count('h1') > 1:
        issues.append("‚ö†Ô∏è Multiple H1 tags found.")
    return issues

def check_image_alt(soup):
    return [img['src'] for img in soup.find_all('img') if not img.get('alt')]

def run_seo_audit(url):
    if not validators.url(url):
        return "üö´ Invalid URL. Please enter a valid URL starting with http or https."

    soup = fetch_page(url)
    if not soup:
        return f"‚ùå Could not fetch the page at {url}."

    meta_issues = check_meta_tags(soup)
    broken = check_broken_links(soup, url)
    heading_issues = check_headings(soup)
    images_missing_alt = check_image_alt(soup)

    output = f"üîç SEO Audit for: {url}\n\n"

    def format_section(title, items):
        if not items:
            return f"‚úÖ {title}: No issues found.\n\n"
        else:
            return f"‚ùó {title}:\n" + "\n".join([f"- {i}" for i in items]) + "\n\n"

    output += format_section("Meta Tag Issues", meta_issues)
    output += format_section("Broken Links", broken)
    output += format_section("Heading Issues", heading_issues)
    output += format_section("Images Missing Alt", images_missing_alt)

    return output

# üéõÔ∏è Gradio UI
gr.Interface(
    fn=run_seo_audit,
    inputs=gr.Textbox(label="Enter Website URL", placeholder="https://example.com"),
    outputs=gr.Textbox(label="SEO Audit Report", lines=20),
    title="üîç Technical SEO Audit Tool",
    description="Paste a URL and get a quick SEO health check ‚Äî titles, descriptions, broken links, H1s, and missing alt tags."
).launch()
```
